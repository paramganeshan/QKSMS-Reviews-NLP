## Task 2: PREPROCESS YOUR TEXT

### Code

```python
import string
def remove_punctuation(review):
    review = review.translate(str.maketrans('', '', string.punctuation))
    return review

import re
def remove_special_chars(review):
    review = re.sub(r"[^a-zA-Z0-9 ]", "", review)
    return review

from num2words import num2words
import re
def num_to_word(review):
    numbers = re.findall('[0-9]+', review)
    for num in numbers:
        num_to_int = int(num)
        review = review.replace(num, num2words(num_to_int))
    return review

def remove_extra_spaces(review):
    review = ' '.join(review.split())
    return review

def text_to_lowercase(review):
    review = review.lower()
    return review

import nltk

from nltk.corpus import stopwords
def remove_stopwords(review):
    cachedStopWords = stopwords.words("english")
    review = ' '.join([word for word in review.split() if word not in cachedStopWords])
    return review

# Code from https://www.machinelearningplus.com/nlp/lemmatization-examples-python/
import spacy
def lemmatize(review):
    nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
    doc = nlp(review)
    review = " ".join([token.lemma_ for token in doc])
    return review

# Puts all previous preprocessing into one function
def preprocess_review(review):
    review = remove_punctuation(review)
    review = remove_special_chars(review)
    review = num_to_word(review)
    review = remove_extra_spaces(review)
    review = text_to_lowercase(review)
    review = remove_stopwords(review)
    #review = lemmatize(review)
    return review
```

### Using VADER for calculating Reviews Sentiment

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def vader_sen(path):
    df = pd.read_csv(path)
    data = []
    for index, row in df.iterrows():
        app_name = row[1]
        review = row[3]
        polarity = SentimentIntensityAnalyzer().polarity_scores(row[3])
        data.append([app_name, review, polarity])
    
    df_sentiment = pd.DataFrame(data, columns = ['App’s package name', 'Review', 'Polarity'])
    return df_sentiment
```
